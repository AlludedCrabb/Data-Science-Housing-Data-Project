This blog is about the different visualizations that can come up during a project. We had an assignment to do multivariable regression to find the best way to predict housing prices. After looking at the dataset and correlations, we started with a single predictor square footage. Below is a simple scatter plot. Each dot represents a house or home with an x value of the home's square footage and a y value of the price, these plots are good for quickly establishing wether two variables have a correlation (or relationship). Over this plot is a line plot, which is exactly what it sounds like, a line! This line shows our first attempts at predicting house prices. But what's missing from these plots? Labels! I shouldn't have to explain the graph to you, it should explain itself.

First simple regression model.png

After some consideration of this first attempt, we moved on to trimming the outliers. Any home that had more than 7999 square feet or had a price of more than 44,999,999 was cut from our model. This time for our results we added explanitory labels, made it a bit bigger, and lowered the alpha of each point, that way it's easier to differentiate the clusters of points and the single points. Now it's easy to see that the amount of homes with less than 3000 square feet are far more common than homes with more than 4000 square feet. It's also easier to see that the majority of homes are sold for under 1,000,000 dollars

Simple regression outliers removed.png

This ended our simple regression. We moved on to preparing for multivariable regression. The first step was choosing what predictors to use, and our first step was to pull up a scatter matrix of the dataframe we'd been using to look for linear relationships between predictors. This plot is big and full of all kinds of information. lets break it down a little. each square represents the relationship between the column and the row, for instance the top square on the far left shows the relationship between the price row and the price column. Since both of these come from the price variable, the relationship is just a count of the different values of price, this kind of plot is called a histogram, and it's probably very familiar. The square just to the right of that one shows the relationship between price and the number of bedrooms in the home. This is a scatter plot like the ones we saw above. a quick look at the bedrooms column shows us that bedrooms rarely ever has a value higher than 10, but it's hard to make conclusions about the relationship between bedrooms and price because the scatter plot is compacted to the left side because of the limited values of bedrooms.

project scatter matrix.png

When you're dealing with a lot of features as potential predictors for your regression, it can be far easier to use another plot called a heatmap. This plot shows the relationships as either light or dark colors. light if the relationship is positive, or dark if the relationship is negative. This is displaying all of the infomation we saw in the scatter matrix, but in a much smaller space and with more clarity. price is still a perfect relationship with itself, so the top left square is very bright. Each feature is perfectly correlated to itself, so the intersection of each column with it's respective row will always be the brightest color on the heatmap.

Now that we have a working model, we decided to see how well it performs. we noticed that the model works better inside Seattle's city limits than in the surrounding areas, but we wanted to know how differently it performs. so we used sql commands with pandas to split up our data by zipcode, then we tested our model on each zipcode individually and plotted those results to a bar graph. a bar graph is different from a histogram in that it shows the relationship between two variables, histograms only show counts. As you can see from the bar chart below, we have just a little variation in results. Some of our zipcodes have a pretty low mean squared error, but some have more than double the mean squared error. looks like our model needs some more work.
